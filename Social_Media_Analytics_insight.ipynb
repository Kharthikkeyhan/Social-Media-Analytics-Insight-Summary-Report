# **Data Cleaning**


```python
# 1. Detect Missing Values

print("Missing Values:")
print(df.isnull().sum())
```

```python

print("\nUsing isna():")
print(df.isna().sum())
```

```python
##Handling missing values
# Numerical columns - fill with median
num_cols = ['likes', 'comments', 'shares', 'follower_count']

for col in num_cols:
  df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.mean()))
```

```python
 df[num_cols]
```

```python
# Categorical columns - fill with mode
cat_cols = ['gender', 'sentiment', 'post_type']
for col in cat_cols:
  df[cat_cols] = df[cat_cols].transform(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
df[cat_cols]
```

```python
# Forward Fill
df = df.ffill()
```

```python
# Backward Fill
df = df.bfill()
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Count and display the total number of duplicate rows

## duplicate handling

duplicates = df.duplicated().sum()
print("\nTotal Duplicate Rows:", duplicates)

```

```python
# Remove duplicates
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

#**Data Formatting**

```python
# Fix incorrect data types

df['likes'] = pd.to_numeric(df['likes'], errors='coerce')
df['comments'] = pd.to_numeric(df['comments'], errors='coerce')
df['shares'] = pd.to_numeric(df['shares'], errors='coerce')
```

```python
# Convert date column
df['posted_at'] = pd.to_datetime(df['posted_at'], errors='coerce')
```

```python
# Standardize Categories

df['gender'] = df['gender'].str.lower().str.strip()

df['gender'] = df['gender'].replace({
    'm': 'male',
    'male ': 'male',
    'f': 'female',
    'female ': 'female'
})

print("\nGender Categories:")
print(df['gender'].unique())
```

```python
##Correct unrealistic values in likes, comments, shares

# Remove negative values

df.loc[df['likes'] < 0, 'likes'] = 0
df.loc[df['comments'] < 0, 'comments'] = 0
df.loc[df['shares'] < 0, 'shares'] = 0
```

```python
# Optional: Cap very large values
df['likes'] = df['likes'].clip(upper=df['likes'].quantile(0.99))
df['comments'] = df['comments'].clip(upper=df['comments'].quantile(0.99))
df['shares'] = df['shares'].clip(upper=df['shares'].quantile(0.99))
```

#**Feature cleaning**

```python
# Extract Hashtag Count

df['hashtag_count'] = df['hashtags'].str.count('#')

print("\nHashtag Count:")
print(df['hashtag_count'].value_counts())
```

```python
# Clean Sentiment Labels

df['sentiment'] = df['sentiment'].str.lower().str.strip()

df['sentiment'] = df['sentiment'].replace({
    'positive ': 'positive',
    'negative ': 'negative',
    'neutral ': 'neutral'
})

print("\nSentiment Categories:")
print(df['sentiment'].unique())
```

```python
# Final Dataset Check
print("\nCleaned Dataset Shape:", df.shape)
print(df.head())
```

#**Data Wrangling**

```python
# Create Engagement Score

df['engagement_score'] = (
    df['likes'] +
    df['comments'] +
    df['shares']
)
```

```python
df['engagement_score']
```

```python
# Log Transformation


df['log_likes'] = np.log1p(df['likes'])

df['log_followers'] = np.log1p(df['follower_count'])
```

```python
# GroupBy Summary - Post Type

post_summary = df.groupby('post_type')[
    ['likes', 'comments', 'shares']
].mean()

print("\nPost Type Summary:")
print(post_summary)
```

```python
# GroupBy Summary - Country

country_summary = df.groupby('country')[
    ['impression_count', 'engagement_score']
].mean()

print("\nCountry Summary:")
print(country_summary)
```

```python
# GroupBy Summary - Sentiment

sentiment_summary = df.groupby('sentiment')[
    ['likes', 'engagement_score']
].mean()

print("\nSentiment Summary:")
print(sentiment_summary)
```

```python
# TASK 5 - STATISTICAL ANALYSIS

stats_cols = [
    'likes',
    'comments',
    'shares',
    'watch_time',
    'engagement_rate',
    'followers'
]

for col in stats_cols:

    if col in df.columns:

        print(f"\n----- {col} -----")

        print("Mean:", df[col].mean())
        print("Median:", df[col].median())
        print("Mode:", df[col].mode()[0])

        print("Std Dev:", df[col].std())
        print("Variance:", df[col].var())

        print("25th Percentile:",
              df[col].quantile(0.25))

        print("50th Percentile:",
              df[col].quantile(0.50))

        print("75th Percentile:",
              df[col].quantile(0.75))

        print("90th Percentile:",
              df[col].quantile(0.90))

        print("Skewness:",
              df[col].skew())

        print("Kurtosis:",
              df[col].kurt())
```
